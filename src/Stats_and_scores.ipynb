```python
import numpy as np
import pandas as pd
import time

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

import warnings
warnings.filterwarnings('ignore')
```

## Junglers stats and scores

```python
jun_score = pd.read_csv('../data/scores/jun_score.csv')

jun_score.head()
```

```python
jun_stats = pd.read_csv('../data/cuadro_mando/jun_stats_app.csv')

jun_stats.head()
```

```python
jun_stats_score = pd.merge(jun_stats, jun_score, on = 'nombre_invocador')

jun_stats_score.head()
```

```python
jun_stats_score.drop('total', axis = 1, inplace = True)
```

```python
jun_stats_score.to_csv('../data/cuadro_mando/jun_stats_score.csv')
```
