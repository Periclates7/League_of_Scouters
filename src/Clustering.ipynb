```python
import numpy as np
import pandas as pd
import time

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

import warnings
warnings.filterwarnings('ignore')

from sklearn.preprocessing import LabelEncoder
```

```python
from sklearn.linear_model import LogisticRegression            # clasificar, logistic regression model   

from sklearn.preprocessing import StandardScaler               # standarized
from sklearn.preprocessing import LabelEncoder               

from sklearn.model_selection import train_test_split as tts 
```

```python
df = pd.read_csv('../data/total.csv')

df.head()
```

```python
df = df.loc[df['rol_1'] == 'Jungler']
             
df.head()
```

```python
df.nivel.unique()
```

```python
df2 = pd.read_csv('../data/total_prof.csv')
```

```python
df2 = df2.loc[df2.rol_1 == 'Jungler']
```

```python
data = pd.concat([df, df2])

data.shape
```

```python
data.nivel.unique()
```

```python
data.head()
```

```python
data.head()
```

```python
data.nivel.unique()
```

```python
data.drop(['region', 'cola'], axis = 1, inplace = True)
```

```python
data.liga.unique()
```

```python
data.liga = data.liga.apply(lambda x: 1 if x == 'Challenger' else 0)
```

```python
df = data.copy()
```

```python
df = df.drop_duplicates()
```

```python
df = df.reset_index()
```

```python
df.drop('index', axis = 1, inplace = True)
```

```python
le=LabelEncoder()

for c in df.columns:
    if c != 'nombre_invocador' and c != 'nivel':
        if df.dtypes[c]==object:

            le.fit(df[c].astype(str))

            df[c]=le.transform(df[c].astype(str))
```

```python
data_num=pd.DataFrame(StandardScaler().fit_transform(df._get_numeric_data()),
                      
                      columns=df._get_numeric_data().columns)


data_obj=df.select_dtypes(include='object')

data2=pd.concat([data_num, data_obj], axis=1)

data2.head()
```

```python
data2 = data2.drop_duplicates()
```

```python
data2 = data2[~data2.index.duplicated()]
data2 = data2.reindex()
```

```python
le=LabelEncoder()

for c in data2.columns:
    if c != 'nombre_invocador':
        if data2.dtypes[c]==object:

            le.fit(data2[c].astype(str))

            data2[c]=le.transform(data2[c].astype(str))
```

```python
data2.head()
```

```python
data2.drop('nombre_invocador', axis = 1, inplace = True)
```

```python
X = data2.drop('nivel', axis = 1)
y = data2.nivel
```

```python
X_train, X_test, y_train, y_test=tts(X, y, random_state=42, stratify=y)
```

```python
logreg=LogisticRegression(max_iter=2000)
```

```python
logreg.fit(X_train, y_train)
```

```python
y_pred=logreg.predict(X_test)

y_pred[:10]
```

```python
set(y_pred)
```

```python
y_prob=logreg.predict_proba(X_test)

y_prob[:10]
```
