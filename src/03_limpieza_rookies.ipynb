# Limpieza dataset de rookies
  
En este archivo se lleva a cabo exactamente el mismo proceso que se ha seguido en el archivo 'limpieza_profesionales.ipynb'. Se recomienda su consulta para entender el proceso que se ha llevado en el presente.  

```python
import numpy as np
import pandas as pd
import time

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

import warnings
warnings.filterwarnings('ignore')
```

```python
from pymongo import MongoClient
cursor=MongoClient()
final_proj = cursor.lol_scouting                                                   
colec = final_proj.lol_scouting 
```

```python
data = [e for e in colec.find()]
```

```python
df = pd.DataFrame(data)

df.head()
```

```python
df.info()
```

```python
df.drop('_id', axis = 1, inplace = True)
```

```python
def transformar_ranking(x):
    x = x.split(' ')
    return x[1][:-1]
```

```python
df.ranking_porcentaje = df.ranking_porcentaje.apply(transformar_ranking)
```

```python
df.head()
```

```python
def quitar_coma(x):
    return x.replace(',', '')
```

```python
df.ranking_mundial = df.ranking_mundial.apply(quitar_coma)
```

```python
ind = df.loc[df['lps'] == ''].index
```

```python
df.drop(index = ind, inplace = True)
```

```python
df.lps = df.lps.astype(int)
```

```python
df.ranking_mundial = df.ranking_mundial.astype(int)
```

```python
df.ranking_porcentaje = df.ranking_porcentaje.astype(float)
```

```python
df.wins = df.wins.astype(int)
```

```python
df.loses = df.loses.astype(int)
```

```python
df.n_games = df.n_games.astype(int)
```

```python
df.winrate_total = df.winrate_total.astype(float)
```

```python
df.liga.unique()
```

```python
df['liga_label'] = df.liga.apply(lambda x: 2 if x == 'Challenger' else 1)
```

```python
df.head()
```

```python
df.drop_duplicates(inplace = True)
```

```python
df.info()
```

```python
df.to_csv('../data/info_general.csv', index = False)
```
