# Filtración de DataFrames
  
En este archivo de texto conformaremos los DataFrames definitivos para su implantación en el cuadro de mando de Streamlit.

```python
# Carga de recursos y .csv
```

```python
import numpy as np
import pandas as pd
import time

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

import warnings
warnings.filterwarnings('ignore')
```

```python
total = pd.read_csv('../data/total.csv')

total.head()
```

```python
total_prof = pd.read_csv('../data/total_prof.csv')

total_prof.head()
```

```python
# Eliminación de los jugadores profesionales que aparecen en el df de rookies.
```

```python
total = total[~total['nombre_invocador'].isin(total_prof['nombre_invocador'].unique())]
```

```python
total_total = pd.concat([total_prof, total])
```

```python
# Concatenamos ambos dfs
```

```python
total_total.to_csv('../data/concat_prof_rook.csv')
```

```python
total_total.shape
```

## Construcción dfs app (Jungler)

```python
# Este proceso se repetirá 5 veces para cada uno de los roles
```

```python
# Filtramos nuestro df para quedarnos con los jugadores del rol que nos interesa
```

```python
total_jun = total_total.loc[total_total.rol_1 == 'Jungler']
```

```python
total_jun.head()
```

```python
# Cargamos el df análogo con los scores conseguidos
```

```python
jun_score = pd.read_csv('../data/scores/jun_score.csv')

jun_score.head()
```

```python
# Nos quedamos con los valores que nos interesan para hacer merge con el principal
```

```python
jun_score = jun_score[['nombre_invocador', 'total']]
```

```python
total_jun = pd.merge(total_jun, jun_score, on = 'nombre_invocador')
```

```python
total_jun.head()
```

```python
# Ordenamos de mayor a menor para obtener un ranking
```

```python
total_jun = total_jun.sort_values(by = 'total', ascending = False)
```

```python
total_jun.head()
```

```python
# Renombro la columna 'total'
```

```python
total_jun = total_jun.rename(columns = {'total':'score'})
```

```python
# Filtro el df original para quedarme solo con los datos referentes a estadísticas generales de cada jugador
```

```python
jun_stats_app = total_jun[['nombre_invocador', 'region', 'liga', 'lps', 'ranking_mundial', 'winrate_total', 'kills_player_kda', 'deaths_player_kda', 'assists_player_kda', 'score']]

jun_stats_app.head()
```

```python
jun_stats_app = jun_stats_app.fillna('EUW')
```

```python
jun_stats_app = jun_stats_app.drop_duplicates()
```

```python
# Exporto a .csv el primer df
```

```python
jun_stats_app.to_csv('../data/cuadro_mando/jun_stats_app.csv', index = False)
```

```python
# Filtro por los campeones más jugados por cada jugador con sus correspondientes estadísticas.
```

```python
jun_champ_app = total_jun[['nombre_invocador', 'champ_1', 'regional_rank_1', 'kills_champ_1', 'deaths_champ_1', 'assists_champ_1', 'winrate_champ_1', 'champ_2', 'regional_rank_2', 'kills_champ_2', 'deaths_champ_2', 'assists_champ_2', 'winrate_champ_2', 'champ_3', 'regional_rank_3', 'kills_champ_3', 'deaths_champ_3', 'assists_champ_3', 'winrate_champ_3', 'score']]

jun_champ_app.head()
```

```python
jun_champ_app = jun_champ_app.drop_duplicates()
```

```python
# Exporto el segundo df
```

```python
jun_champ_app.to_csv('../data/cuadro_mando/jun_champ_app.csv', index = False)
```
